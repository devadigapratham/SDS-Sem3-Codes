```python
# demonstration of the central limit theorem
# as sample sizes becomes larger
# distribution of sample mean's -> normal distribution
from numpy.random import seed
from numpy.random import randint
from numpy import mean
from matplotlib import pyplot
# seed -> initialize random generator
# saves the state of random function
```

```python
# how seed works
print("without seed fun,1st try",randint(0,50,3))
print("without seed fun,2nd try",randint(0,50,3))
seed(2)
print("with seed=2,1st try",randint(0,60,3))
seed(2)
print("with seed=2,2nd try",randint(0,60,3))
```

```python
# seed the random number generator
seed(1)

def plot_clt(n):# n-> no.of samples
    # calculate the mean of 50 dice rolls n times
    # randint(start,stop,no.of integers/sample size)
    means = [mean(randint(1, 7, 50)) for _ in range(n)]

    # plot the distribution of sample means
    pyplot.hist(means)
    pyplot.show()
# as n(no.of samples)increses distribution of sample mean's becomes normal distribution
plot_clt(100)# 100=no.of samples
plot_clt(1000)
plot_clt(100000)
```

```python
# Z-score : how many stdevs a data point is from the mean of a distribution
# Z-score for given area

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

area = 0.95     #area under the graph
# convert values to z-score
print('z-score:',norm.ppf(area))
# ppf -> percent point function
# ppf(x) -> inverse of CDF(cumulative distribution function)
# x value for probability(area)
# ppf => area or probability to z-score
# cdf => z-score to area or probability
print('area :',norm.cdf(1.6448536269514722))#left side area
```

```python
# z-score to value/area
print(norm.cdf(1.6448536269514722))
```

```python
#Given X ~ N(mu, var)
#Finding Probabilities

import matplotlib.pyplot as plt
#import matplotlib as plt
import numpy as np
from scipy.stats import norm

mu = 40 # population mean
x = 47 # raw score/data point
sigma = 8 # population st dev
z = (x - mu)/sigma
print('Finding P(z <',z,')')
print('Area to left of x : ',norm.cdf(z))

print('Finding P(z >', z,')')
print('Area to right of x : ',1 - norm.cdf(z))
#since total area under the graph is 1
```

```python
#Area in an interval between two values a and b
from scipy.stats import norm
def prob(a,b):# a,b are datapoints
    z1 = (a - mu)/sigma
    z2 = (b - mu)/sigma
    print('Finding P(',z1,'< z <', z2,')')
    print('Area to left of a : ',norm.cdf(z1))
    print('Area to left of b : ',norm.cdf(z2))
    print('Area to between a and b: ',norm.cdf(z2) - norm.cdf(z1))
    
mu = 50
sigma = 5
prob(44.5, 55.5)   
```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from scipy.stats import norm

# example data
mu = 100 # mean of distribution
sigma = 15 # standard deviation of distribution
x = mu + sigma * np.random.randn(10000)# size of x is 10000

num_bins = 20
# the histogram of the data
n, bins, patches = plt.hist(x, num_bins,density=1, facecolor='blue', alpha=0.5)#alpha->transparacy

# add a 'best fit' line
y = norm.pdf(bins, mu, sigma)# probability density function
plt.plot(bins, y, 'r--')
plt.xlabel('bins')
plt.ylabel('Probability')
plt.title(r'Histogram of IQ: mean=100, sigma=15')

# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
plt.show()
```
