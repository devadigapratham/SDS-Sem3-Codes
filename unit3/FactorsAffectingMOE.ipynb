```python
#Effect of Standard deviation on Margin of error
#MOE = z_alpha * sigma/sqrt(n)
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.stats import norm

#sigma = 10
z_score = 1.96 # 95% CL
n = 64      

moe = []  #list of margin of errors
x = range(0,21) # Varying SD from 0 to 20

for sd in x:
    moe.append(z_score * sd/math.sqrt(n))
    
plt.plot(x, moe)
plt.title('n = 64, CL = 95% and Varying the Standard deviation')
plt.ylabel('Margin of error')
plt.xlabel('Standard deviation')
plt.show()

```

```python
#Effect of Confidence Level on Margin of error
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.stats import norm


sigma = 10
#z_score = 1.96 # 95% CL
n = 64

# probability that random variable from std normal dist falls btw z1,z2
def prob(z1,z2):# or confidence level
       return (norm.cdf(z2) - norm.cdf(z1))
 
# array of z-scores
z_alphaby2 = np.linspace(norm.ppf(0.10), norm.ppf(0.0005), 100)#creates evenly spaced 100 datapoints
# print(z_alphaby2)
z_alphaby2 = z_alphaby2 * -1 # as 0.1 and 0.0005 is less than 0.5;if its greater than 0.5 no need of *-1
# print(z_alphaby2)

#calculating margin of error
moe = z_alphaby2 * sigma/math.sqrt(n)

#getting the CL
x = []
for i in range(0, len(z_alphaby2)):
    x.append(prob(-1 * z_alphaby2[i], z_alphaby2[i]))

#getting the CL in %
x = np.array(x) * 100

plt.plot(x, moe)
plt.ylabel('Margin of error')
plt.ylim(0.5,5)
plt.xlim(80,100)
plt.title('sigma = 10, n = 64 and Varying the CL')
plt.xlabel('confidence level')
plt.show()
```
